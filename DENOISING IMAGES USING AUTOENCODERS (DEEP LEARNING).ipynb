# Denoising Images Through Deep Learning Autoencoders

`Encode` means _to convert into a particular or different form_. Also, we can add that this new form can be tipically a simpler or reduced form. Then, we could came into conclusion that an **Autoencoder** is a deep learning model which is capable to _"teach"_ itself how to encode information in a simpler form, which it was provided, in the sense of reduced dimentions that the original input.

## Layers in an Autoencoder
It can be notice that the number of nodes in the `input layer` is the same as in the `output layer`.  
Let's say that the number of nodes in the the input or output layer is $K$, the number of nodes in the encoder or decoder layer is $N$, and the number of nodes in the code layer is $M$, then we can state that $K>N>M$, so the code layer is a representation of the input in a reduced for and the output is a `construction` from this simpler form into a equivalent of the original input.

**So, the goal of an autoencoder is to get the output to match the input as closely as possible.**


## Use of Autoencoders

There are several applications for Autoencoders, where the most relevant are:
- Data Compression.
- Reduction of Dimenssions.
- Data Cleaning.
- Feature Exraction.
- Anomaly Detection.

In this example, we going to explore the use of an Autoencoder to denoise images which get into the _Data Cleaning_ category which is a critical step for any Machine Learning pipeline.

# The Experiment

The [MNIST dataset](https://en.wikipedia.org/wiki/MNIST_database) which is a set of images of handwritten digits is used to train an Autoencoder. Then, we're going to introduce `noise` into the test images by occlusion of a section of the image in order to test the ability of the Autoencoder to remove the `noise`, reconstructing the image using the reduced interpretation learned previously.


```python
import numpy as np
import matplotlib.pyplot as plt
import copy
from time import time

import torch
import torch.nn as nn
import torch.nn.functional as F

%matplotlib inline
```

```python
# load dataset
# Notice that we don't need labels due to the Autoencoder learn from the input examples.
train = np.load('./Data/X_train.npy')
test = np.load('./Data/X_test.npy')

# normalize the data to a range of [0 1]
trainNorm = train[:20000] / np.max(train[:20000]) # Only the first 20,000 of 60,000 images are used.
testNorm = test / np.max(test)

# convert to tensor
T_train = torch.tensor( trainNorm ).float()
T_test = torch.tensor( testNorm ).float()
```

## Example of noise insertion by occlusion

```python
# images took from train dataset
dig1 = T_train[1000,:].view(28,28)
occl1 = copy.deepcopy(dig1)
occl1[10:13,:] = 1

dig2 = T_train[7000,:].view(28,28)
occl2 = copy.deepcopy(dig2)
occl2[:,7:10] = 1
fig,ax = plt.subplots(2,2,figsize=(8,5))

ax[0,0].imshow(dig1,cmap='gray')
ax[0,0].set_title('Original image')
ax[0,0].axis('off')

ax[0,1].imshow(occl1,cmap='gray')
ax[0,1].set_title('Occluded image')
ax[0,1].axis('off')

ax[1,0].imshow(dig2,cmap='gray')
ax[1,0].set_title('Original image')
ax[1,0].axis('off')

ax[1,1].imshow(occl2,cmap='gray')
ax[1,1].set_title('Occluded image')
ax[1,1].axis('off')

plt.show()
```

## Deep Learning Model

```python
def create_autoencoder():

  class AEnet(nn.Module):
    def __init__(self):
      super().__init__()

      ### input layer
      self.input = nn.Linear(784,128)
      ### encoder layer
      self.enc = nn.Linear(128,50)
      ### Code layer
      self.lat = nn.Linear(50,128)
      ### decoder layer
      self.dec = nn.Linear(128,784)

    # forward pass
    def forward(self,x):
      x = F.relu( self.input(x) )
      x = F.relu( self.enc(x) )
      x = F.relu( self.lat(x) )
      y = torch.sigmoid( self.dec(x) )
      return y
  
  # create the model instance
  net = AEnet()
  
  # loss function
  lossfun = nn.MSELoss()

  # optimizer
  optimizer = torch.optim.Adam(net.parameters(),lr=.001)

  return net,lossfun,optimizer
```

## Train Model Function

```python
def fit_model():

  # number of epochs (each epoch goes through all samples)
  numepochs = 5
  # create a new model
  net,lossfun,optimizer = create_autoencoder()

  # initialize losses
  losses = []

  # batch size and number of batches
  batchsize  = 32
  numBatches = int(T_train.shape[0]/batchsize)

  # loop over epochs 
  for epochi in range(numepochs):

    # get a permuted index vector
    randidx = np.random.permutation(T_train.shape[0]).astype(int)
    
    for batchi in range(numBatches):
      
      # samples to use in this batch
      samps2use = range((batchi-1)*batchsize,batchi*batchsize)

      # select those images
      X = T_train[randidx[samps2use],:]

      # forward pass and loss
      yHat = net(X)
      loss = lossfun(yHat,X)

      # backprop
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

    
      # losses in this batch
      losses.append( loss.item() )

  # end epochs
  
  return losses,net
```

## Model Training

```python
# train the model
start = time()
losses, net = fit_model()
print(f'Final loss: {losses[-1]:.4f}')
print(f'Elapsed time during training: {(time()-start):.2f} seconds')
```

```python
# visualizing the losses
plt.plot(losses, color='orange')
plt.title('Model Loss', fontsize=18)
plt.xlabel('Epochs', fontsize=16)
plt.ylabel('MSE', fontsize=16)
plt.grid()
plt.show()
```

## Add occlusion to test images

```python
X = copy.deepcopy(T_test)

# add noise
for i in range(X.shape[0]):
  
  # reshape the image
  image = X[i,:].view(28,28)

  # occlude random rows or columns
  startloc = np.random.choice(range(10,21))
  if i%2==0: # even -> horizontal occlusion
    image[startloc:startloc+1,:] = 1
  else:      # odd -> vertical occlusion
    image[:,startloc:startloc+1] = 1

```

## Run the noisy images through the model

```python
denoise = net(X)
```
